```python
#imports
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from scipy.interpolate import lagrange
from numpy.polynomial.polynomial import Polynomial
from collections import Counter
import csv
import time
from multiprocessing import Process
import pickle
```

```python
def knn(obj, data, k):
    distDict = {}
    for i in range(len(data)):
        distDict[i] = np.linalg.norm(np.array(data[i][0:len(data[i]) - 1]) - np.array(obj[0:len(data[i]) - 1]))
    nearest = sorted(distDict.items(), key=lambda x:x[1])[0:k]
    labels = []
    for i in nearest:
        labels.append(data[i[0]][-1])
    return Counter(labels).most_common(1)[0][0]    
```

```python
#q1
q1data0 = []
q1data1 = []
with open("D2z.txt") as f:
      for i in f.readlines():
            line = i.split()
            if int(line[2]) == 0:
                q1data0.append([float(line[0]), float(line[1]), int(line[2])])
            else:
                q1data1.append([float(line[0]), float(line[1]), int(line[2])])
q1data = q1data0 + q1data1
q1data0 = np.array(q1data0)
q1data1 = np.array(q1data1)
```

```python
x, y = np.meshgrid(np.arange(-2,2.1,0.1),np.arange(-2,2.1,0.1))
pred = []
for i in range(len(x)):
        for j in range(len(x[0])):
            pt = (x[i][j], y[i][j])
            pred.append(knn(pt, q1data, 1))
pred = np.array(pred).reshape(x.shape)
plt.scatter(x, y, c=pred, marker='.', s=1)
plt.plot(q1data0[:,0], q1data0[:,1], marker='o', color='k', linestyle='none', ms=3)
plt.plot(q1data1[:,0], q1data1[:,1], marker='x', color='k', linestyle='none', ms=3)
plt.show()
```

# Spam Filter

```python
#load data
with open('emails.csv', encoding="utf-8") as f:
    csvData = list(csv.reader(f))
header = csvData[0]
data = csvData[1:]
data = [[int(col) for col in row[1:]] for row in data]
fold1 = (data[0:1000], data[1000:])
fold2 = (data[999:2000], data[0:999]+data[2000:])
fold3 = (data[1999:3000], data[0:1999]+data[3000:])
fold4 = (data[2999:4000], data[0:2999]+data[4000:])
fold5 = (data[3999:], data[0:3999])
q5data = (data[4000:],data[0:4000])
folds = [fold1, fold2, fold3, fold4, fold5]
```

```python
#K-nearest neighbor
knn5fold = {}
for k in [1,3,5,7,10]:
    print("\nk="+str(k))
    allFolds = {}
    for i in range(len(folds)):
        results={"TP":0, "FP": 0, "FN":0, "TN": 0}
        fold = folds[i]
        for j in fold[0]:
            pred = knn(j, fold[1], k)
            if j[-1] == 0:
                if pred == 0:
                    results["TN"] = results["TN"] + 1
                else:
                    results["FP"] = results["FP"] + 1
            else:
                if pred == 0:
                    results["FN"] = results["FN"] + 1
                else:
                    results["TP"] = results["TP"] + 1 
        allFolds[i] = results
        print("fold" + str(i+1))
        print("Accuracy: " + str((results["TP"]+results["TN"])/(results["TP"]+results["TN"]+results["FP"]+results["FN"])))
        print("Precision: " + str(results["TP"]/(results["TP"]+results["FP"])))
        print("Recall: " + str(results["TP"]/(results["TP"]+results["FN"])))
    knn5fold[k] = allFolds
with open('knn5fold.results', 'wb') as f: # Cache results
    pickle.dump(knn5fold, f)
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
# Logistic regression
def gradientDescent(fold):
    theta = np.zeros([1,3001]).T
#     fold = fold1
    data =  np.array(fold[1])
    labels = np.array([data[:,-1]]).T
    a=np.zeros([1,len(data)]).T
    a.fill(1)
    train = np.hstack([a, data[:,0:-1]])
    counter = 0
    while True:
        pred = sigmoid(train@theta)
        diff = pred - labels
        gradient = np.array([(np.array(train[:,col])@diff) / train.shape[0] for col in range(train.shape[1])])
        theta = theta - 0.001 * gradient
        if np.linalg.norm(gradient) < 0.001:
            break
        if counter > 4000:
            break
        counter += 1
#         print(np.linalg.norm(gradient))
    return theta
```

```python
#q3
for i in range(len(folds)):
    fold = folds[i]
    print("fold"+str(i+1))
    theta = gradientDescent(fold)
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for j in fold[0]:
        x = np.array([1] + j[0:-1])
        pred = sigmoid(x@theta)
        if pred > 0.5:
            if j[-1] == 1:
                TP += 1
            else:
                FP += 1
        else:
            if j[-1] == 1:
                FN += 1
            else:
                TN += 1        
    print("Accuracy: " + str((TP+TN)/(TP+TN+FP+FN)))
    print("Precision: " + str(TP/(TP+FP)))
    print("Recall: " + str(TP/(TP+FN)))
```

```python
#q4
kList = []
avgacc = []
for i in knn5fold:
    kList.append(i)
    acctotal = 0
    for j in knn5fold[i]:
        results = knn5fold[i][j]
        acctotal += (results["TP"]+results["TN"])/(results["TP"]+results["TN"]+results["FP"]+results["FN"])
    avgacc.append(acctotal/5)
plt.plot(kList, avgacc, marker='o')
plt.xlabel("k")
plt.ylabel('Average accuracy')
plt.title("kNN 5-Fold Cross Validation")
plt.grid()
plt.show()
print("K: " + str(kList))
print("Average accuracy for each K: " + str(avgacc))
```

```python
#q5
#logistic regression
allpred = []
theta = gradientDescent(q5data)
for i in q5data[0]:
    x = np.array([1] + i[0:-1])
    pred = sigmoid(x@theta)
    allpred.append(pred)
allpred = list(set(np.array(allpred).flatten().tolist()))
fpr = []
tpr = []
for i in allpred:
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for j in q5data[0]:
        x = np.array([1] + j[0:-1])
        pred = sigmoid(x@theta)
        if pred > i:
            if j[-1] == 1:
                TP += 1
            else:
                FP += 1
        else:
            if j[-1] == 1:
                FN += 1
            else:
                TN += 1   
    fpr.append(FP/(FP+TN))
    tpr.append(TP/(TP+FN))
points = []
for i in range(len(fpr)):
    points.append([fpr[i],tpr[i]])
if not [0,0] in points:
    points.append([0,0])
if not [1,1] in points:
    points.append([1,1])
points = sorted(points, key=lambda x:x[0])
points = np.array(points)
#     print(FP/(FP+TN))
#     print(TP/(TP+FN))
# print(fpr)
# print(tpr)
```

```python
def knn2(obj, data, k):
    distDict = {}
    for i in range(len(data)):
#         dist = 0
#         for j in range(len(data[i]) - 1):
#             dist += (data[i][j] - obj[j]) ** 2
#         distDict[i] = dist
        distDict[i] = np.linalg.norm(np.array(data[i][0:len(data[i]) - 1]) - np.array(obj[0:len(data[i]) - 1]))
    nearest = sorted(distDict.items(), key=lambda x:x[1])[0:k]
    labels = []
    for i in nearest:
        labels.append(data[i[0]][-1])
    return Counter(labels)
```

```python
#5 nearest neighbors
allpredknn = []
for i in q5data[0]:
    pred = knn2(i, q5data[1], 5)[1]
    allpredknn.append(pred)
```

```python
val = list(set(np.array(allpredknn).flatten().tolist()))
fprknn = []
tprknn = []
for i in val:
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for j in range(len(allpredknn)):
        pred = allpredknn[j]
        if pred > i:
            if q5data[0][j][-1] == 1:
                TP += 1
            else:
                FP += 1
        else:
            if q5data[0][j][-1] == 1:
                FN += 1
            else:
                TN += 1   
    fprknn.append(FP/(FP+TN))
    tprknn.append(TP/(TP+FN))
pointsknn = []
for i in range(len(fprknn)):
    pointsknn.append([fprknn[i],tprknn[i]])
if not [0,0] in pointsknn:
    pointsknn.append([0,0])
if not [1,1] in pointsknn:
    pointsknn.append([1,1])
pointsknn = sorted(pointsknn, key=lambda x:x[0])
pointsknn = np.array(pointsknn)
```
